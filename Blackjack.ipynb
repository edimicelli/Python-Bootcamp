# I built a Blackjack game!  Just run the code below to play!

```python
import random
```

```python
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen','King','Ace')
values = {'Two':2,'Three':3,'Four':4,'Five':5,'Six':6,'Seven':7,'Eight':8,'Nine':9,'Ten':10,'Jack':10,'Queen':10,'King':10,'Ace':11}
```

```python
class Card:
    
# NEEDS SUITS, RANKS, & VALUES
    
    def __init__(self, suit, rank):
        
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
        
        
    def __str__(self):
        
        return f'{self.rank} of {self.suit}'
```

```python
class Player:
    
# NEED HAND
# NEED MONEY
# NEED BET
    
    def __init__(self, name):
        
        self.name = name
        self.hand = []
        self.wallet = 100
        self.bet = 0
```

```python
class Dealer:
    
    def __init__(self):
        
        self.hand = []
```

```python
class Deck:
    
# HAVE ALL CARDS
# SHUFFLE
# ABLE TO DEAL ONE AT A TIME
    
    def __init__(self):
        
        self.deck_cards = []
        for suit in suits:
            for rank in ranks:
                self.deck_cards.append(Card(suit, rank))
                
                
    def shuffle(self):
        
        random.shuffle(self.deck_cards)
        
        
    def deal_one(self):
        
        return self.deck_cards.pop()
```

```python
name = input("What is your name?")
player = Player(name)
dealer = Dealer()
```

```python
game_on = True

while game_on == True:
    
    game_deck = Deck()
    player.hand = []
    dealer.hand = []
    player_hand_values = []
    dealer_hand_values = []
    
    game_deck.shuffle()

    try:
        print(f"Your wallet is at {player.wallet}")
        valid_bet = False
        while valid_bet == False:
            player.bet = int(input("What is your bet?"))
            if player.bet > 0 and player.bet <= player.wallet:
                valid_bet = True
            else:
                print("Invalid bet amount.")
    except ValueError:
        print("Your bet must be an integer")
    else:
        print(f"Your bet is {player.bet}")

    dealer.hand.append(game_deck.deal_one())
    dealer.hand.append(game_deck.deal_one())
    print(f'Dealer showing {dealer.hand[1]}')
    player.hand.append(game_deck.deal_one())
    player.hand.append(game_deck.deal_one())
    print(f'{player.name} has {player.hand[0]} and {player.hand[1]}')

    player_hand_over = False
    player_hand_values = [player.hand[0].value, player.hand[1].value]
    dealer_hand_over = False
    dealer_hand_values = [dealer.hand[0].value, dealer.hand[1].value]

    round_over = False
    while round_over == False:
        
        while player_hand_over == False:        
            if sum(player_hand_values) > 21:
                ace = False
                all_cards_checked = False
                while ace == False and all_cards_checked == False:
                    for card in range(len(player.hand)):
                        if player.hand[card].rank == 'Ace' and player.hand[card].value == 11:
                            player.hand[card].value = 1
                            player_hand_values[card] = 1
                            ace = True
                        elif card == len(player.hand)-1:
                            all_cards_checked = True
                if all_cards_checked == True and sum(player_hand_values) > 21:
                    print("Bust!")
                    player.wallet -= player.bet
                    round_over = True
                    player_hand_over = True
            elif sum(player_hand_values) == 21:
                if len(player.hand) == 2:
                    print("Blackjack!")
                else:
                    print("You win with 21!")
                player.wallet += player.bet
                round_over = True
                player_hand_over = True
            elif sum(player_hand_values) <= 21:
                move_valid = False
                while move_valid == False:
                    move = input("Hit (H) or Stay (S)?")
                    if move == "H" or move == "h" or move == "S" or move == "s":
                        move_valid = True
                    else:
                        print("Invalid selection")
                if move == "H" or move == "h":
                    new_card = game_deck.deal_one()
                    player.hand.append(new_card)
                    player_hand_values.append(new_card.value)
                    print(f'{new_card}')
                else:
                    player_hand_over = True
        
        if round_over == True:
            break
        print(f"Dealer has {dealer.hand[0]} and {dealer.hand[1]}")
        while dealer_hand_over == False:
            if sum(dealer_hand_values) > 21:
                ace = False
                all_cards_checked = False
                while ace == False and all_cards_checked == False:
                    for card in range(len(dealer.hand)):
                        if dealer.hand[card].rank == 'Ace' and dealer.hand[card].value == 11:
                            dealer.hand[card].value = 1
                            dealer_hand_values[card] = 1
                            ace = True
                        elif card == len(dealer.hand)-1:
                            all_cards_checked = True
                if all_cards_checked == True and sum(dealer_hand_values) > 21:
                    print("Dealer busts!")
                    player.wallet += player.bet
                    round_over = True
                    dealer_hand_over = True
            elif sum(dealer_hand_values) == 21:
                if len(dealer.hand) == 2:
                    print("Dealer has Blackjack!")
                else:
                    print("Dealer wins with 21!")
                player.wallet -= player.bet
                round_over = True
                dealer_hand_over = True
            elif sum(dealer_hand_values) <= sum(player_hand_values):
                print("Dealer hits")
                new_card = game_deck.deal_one()
                dealer.hand.append(new_card)
                dealer_hand_values.append(new_card.value)
                print(f'{new_card}')
            else:
                print(f"Dealer wins with {sum(dealer_hand_values)}")
                player.wallet -= player.bet
                round_over = True
                dealer_hand_over = True
            

    if player.wallet == 0:
        print("You're out of money!")
        game_on = False
        
    valid_response = False
    while valid_response == False:
        continue_playing = input("Would you like to continue playing (Y/N)?\n\n")
        if continue_playing == "N" or continue_playing == "n":
            game_on = False
            valid_response = True
        elif continue_playing == "Y" or continue_playing == "y":
            valid_response = True
```
